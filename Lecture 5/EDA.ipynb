```python
import pandas as pd

# Ganti path dengan path sesuai dengan lokasi file dataset Anda di Google Drive
file_path = '/content/drive/MyDrive/Dataset/IT Salary Survey EU  2020.csv'

# Membaca dataset sebagai DataFrame
df = pd.read_csv(file_path)

# Menampilkan lima baris pertama dari DataFrame
print(df.head())
```

```python
# Menampilkan informasi ringkas dari DataFrame
print(df.info())
```

```python
# Mendapatkan summary statistik dari DataFrame
summary_statistic = df.describe()

# Menampilkan summary statistik
print(summary_statistic)
```

```python
# Memilih subset DataFrame hanya dengan fitur-fitur yang diminta
selected_features = df[["Total years of experience",
                        "Years of experience in Germany",
                        "Yearly bonus + stocks in EUR",
                        "Annual bonus+stocks one year ago. Only answer if staying in same country",
                        "Number of vacation days",
                        "Сontract duration",
                        "Company size"]]

# Menampilkan 5 baris teratas dari fitur-fitur tersebut
print(selected_features.head())
```

```python
import pandas as pd

# Kolom-kolom yang ingin diubah menjadi tipe data numerik
columns_to_convert = ["Total years of experience",
                      "Years of experience in Germany",
                      "Yearly bonus + stocks in EUR",
                      "Annual bonus+stocks one year ago. Only answer if staying in same country",
                      "Number of vacation days",
                      "Сontract duration",
                      "Company size"]

# Mengonversi kolom-kolom menjadi tipe data numerik
df[columns_to_convert] = df[columns_to_convert].apply(pd.to_numeric, errors='coerce')

# Menampilkan informasi tentang tipe data setelah konversi
print(df[columns_to_convert].info())
```

```python
# Menampilkan baris yang memiliki nilai null
null_rows = df[df.isnull().any(axis=1)]

# Menampilkan baris yang memiliki nilai null
print(null_rows)
```

```python
# Mengisi nilai null dengan nilai "0" dalam kolom yang dimaksud
df['Have you been forced to have a shorter working week (Kurzarbeit)? If yes, how many hours per week'] = df['Have you been forced to have a shorter working week (Kurzarbeit)? If yes, how many hours per week'].fillna(0)

# Menampilkan baris yang memiliki nilai null setelah pengisian
null_rows_after_filling = df[df.isnull().any(axis=1)]
print(null_rows_after_filling)
```

```python
import matplotlib.pyplot as plt

# Filter data sesuai dengan rentang umur yang diminta
filtered_data = df[(df['Age'] >= 20) & (df['Age'] <= 69)]

# Menghitung jumlah individu di setiap kategori umur
age_counts = filtered_data['Age'].value_counts()

# Membuat pie chart
plt.figure(figsize=(8, 8))
plt.pie(age_counts, labels=age_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribusi Umur (20-69 tahun)')
plt.axis('equal')  # Membuat lingkaran menjadi proporsional
plt.show()
```
