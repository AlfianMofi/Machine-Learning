```python
import pandas as pd

# Membaca file CSV dan membuat DataFrame
df = pd.read_csv('/content/drive/MyDrive/Dataset/vgsales.csv')

# Menampilkan beberapa baris pertama dari DataFrame
print(df.head())
```

# Feature Statistic

```python
# Menampilkan statistik deskriptif untuk fitur-fitur dalam DataFrame
(df.describe())
```

# Box Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Mengatur gaya seaborn untuk plot
sns.set(style="whitegrid")

# Menampilkan box plot untuk setiap fitur numerik
numerical_features = ['Rank', 'Year', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=feature, data=df, orient='v')
    plt.title('Box Plot of ' + feature)
    plt.show()
```

# Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt



# Memilih hanya fitur-fitur numerik
numerical_features = ['Rank', 'Year', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']

# Membuat pairplot untuk fitur-fitur numerik
sns.pairplot(df[numerical_features])
plt.show()
```

# Heat Map

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Memilih hanya fitur-fitur numerik
numerical_features = ['Rank', 'Year', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']

# Menghitung korelasi antara fitur-fitur numerik
correlation_matrix = df[numerical_features].corr()

# Membuat heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
